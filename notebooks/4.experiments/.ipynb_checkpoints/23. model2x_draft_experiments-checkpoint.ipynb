```python
import pandas as pd
```

```python
PATH_TO_DATA = '/media/andrii/earth/Katia/CS_MasterThesis/data/'
PATH_TO_DATA_RL = '/media/andrii/earth/Katia/CS_MasterThesis/data/ukwiki/uk_red_links/'
```

```python
# read a train sample with features
train_sample = pd.read_csv(PATH_TO_DATA_RL+'train_red_links_uk_with_features.csv', encoding = 'UTF-8')
```

```python
# read ground truth
ground_truth = pd.read_csv(PATH_TO_DATA_RL+'ukredlinks_ground_truth_3194.csv', encoding = 'UTF-8')
```

```python
ground_truth_train = pd.merge(train_sample, ground_truth, how='left', left_on='red_link_name', right_on='red link name')
```

```python
ground_truth_train = ground_truth_train[['red_link_name', 'candidate', 'BN_results', 'jaccard_score', 'levenstein_score', 'ground truth']]
```

```python
gt_results = []
for index, row in ground_truth_train.iterrows():
    gt_result = row['candidate'] == row['ground truth']
    gt_results.append(gt_result)
```

```python
gt_df = pd.DataFrame(gt_results)
```

```python
with_gt_df = pd.concat([ground_truth_train, gt_df], axis=1)
```

### Model 1

```python
def BabelNet_model(predictions):
    if predictions == 1:
        return True
    else:
        return False        
```

```python
with_gt_df['BabelNet_model_results'] = train_sample.apply (lambda row: BabelNet_model(row['BN_results']),axis=1)
```

### EVALUATION

```python
def evaluate_BabelNet_model(ground_truth, model_results):
    if ground_truth == model_results == False:
        return 'TN'
    if ground_truth == False and model_results == True:
        return 'FP'
    if ground_truth == model_results == True:
        return 'TP'
    if ground_truth == True and model_results == False:
        return 'FN'
```

```python
def f1_score(evaluation_results):
    res = evaluation_results.value_counts()
    try:
        tp = res['TP']
    except:
        tp = 0
    try:
        tn = res['TN']
    except:
        tn = 0
    try:
        fp = res['FP']
    except:
        fp = 0
    try:
        fn = res['FN']
    except:
        fn = 0
    precision = tp/(tp + fp)
    print('precision =', precision)

    recall = tp/(tp + fn)
    print('recall =', recall)

    f1_score = 2*precision*recall/(precision+recall)
    print('f1_score =', f1_score)
    
    return precision, recall, f1_score
```

```python
with_gt_df['BabelNet_model_evaluation'] = with_gt_df.apply\
(lambda row: evaluate_BabelNet_model(row[0], row['BabelNet_model_results']),axis=1)
```

```python
f1_score(with_gt_df['BabelNet_model_evaluation'])
```
