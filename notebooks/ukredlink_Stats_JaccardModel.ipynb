```python
import pandas as pd
```

```python
bn_results_df = pd.read_csv('uk_red_links_results_lighter.csv')
```

```python
bn_results_df
```

```python
res = bn_results_df['Evaluation'].value_counts()
```

```python
res
```

```python
tp = res['TP']
tn = res['TN']
fp = res['FP']
fn = res['FN']
```

```python
precision = tp/(tp + fp)
print('precision =', precision)
```

```python
recall = tp/(tp + fn)
print('recall =', recall)
```

```python
f1_score = (2*precision*recall)/(precision+recall)
print('f1_score =', f1_score)
```

```python
# with TP in top 5
top5 = bn_results_df['Remarks'].value_counts()
```

```python
top5
```

```python
top5_TP = top5['Right item present in the list']
print(top5_TP)
```

```python
precision_top5 = (tp+top5_TP)/(tp+top5_TP + fp)
print('precision_top5 =', precision_top5)
```

```python
recall_top5 = (tp+top5_TP)/(tp+top5_TP + fn)
print('recall_top5 =', recall_top5)
```

```python
f1_score_top5 = (2*precision_top5*recall_top5)/(precision_top5+recall_top5)
print('f1_score_top5 =', f1_score_top5)
```
