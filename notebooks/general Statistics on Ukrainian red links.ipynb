```python
# check for scripts
from alphabet_detector import AlphabetDetector
ad = AlphabetDetector()
```

```python
ad.detect_alphabet(u's汉字??')
```

```python
def get_scripts(x):
    x_set = ad.detect_alphabet(str(x))
    x_list = list(x_set)
    x_string = ' '.join(x_list)
    return x_string
```
