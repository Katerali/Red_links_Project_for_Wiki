```python
def replace_nans(red_link_name, link_txt):
    if pd.isna(link_txt):
        link_txt = red_link_name
    return link_txt
```

```python
red_link_txt['link_text'] = red_link_txt.apply(lambda row: replace_nans(row['red_link_name'], row['link_txt']),axis=1)
```

```python
flist = open(PATH_TO_DATA_TEXT+"wiki_00.txt").readlines()

parsing = False

itercars = iter(flist)
for line in flist:
    if line.startswith('<doc id="12"'):
        parsing = True
    if parsing:
        if 'political philosophy' in line:
            line_list = line.split()
            line_list_cleaned = []
            for r in line_list:
                if not r in stop_words:
                    line_list_cleaned.append(r)
            res = line_list_cleaned[:20]
            print(res)
        next(itercars)
        if line.startswith('<doc id='):
            parsing = False
```
