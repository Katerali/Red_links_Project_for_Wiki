```python
df_proper_names = bn_results_df[bn_results_df['Proper Name']=='PN']
```

```python
all_pn = len(df_proper_names)
print('all_pn =', all_pn)

res_pn = df_proper_names['Evaluation'].value_counts()
print(res_pn)

tp = res_pn['TP']
tn = res_pn['TN']
fp = res_pn['FP']
fn = res_pn['FN']

precision = tp/(tp + fp)
print('precision =', precision)

recall = tp/(tp + fn)
print('recall =', recall)

f1_score = (2*precision*recall)/(precision+recall)
print('f1_score =', f1_score)
```

```python
df_latin = bn_results_df[bn_results_df['Latin']=='Latin']
```

```python
all_lat = len(df_latin)
print('all_lat =', all_lat)

res_lat = df_latin['Evaluation'].value_counts()
print(res_lat)

tp = res_lat['TP']
#tn = res_lat['TN']
fp = res_lat['FP']
#fn = res_lat['FN']

precision = tp/(tp + fp)
print('precision =', precision)

recall = tp/(tp + 0)
print('recall =', recall)

f1_score = (2*precision*recall)/(precision+recall)
print('f1_score =', f1_score)
```
