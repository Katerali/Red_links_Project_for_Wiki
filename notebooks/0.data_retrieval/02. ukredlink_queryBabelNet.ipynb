```python
total_list_of_results = []
f1=open('./requests_results_wikidata.txt', 'w+')
for link in list_of_red_links:
    url = "https://babelnet.io/v5/getSenses?lemma=" + link + "&searchLang=UK&targetLang=EN&source=WIKI&key="+myKey
    response = requests.get(url)
    f1.write(response.text + "\n\n")
    json_response = response.json()
    if json_response == []:
        total_list_of_results.append(defaultdict(list))
    else:
        dd = defaultdict(list)
        row = []
        for i in json_response:
            found = {}
            lemma = i['properties']['fullLemma']
            language = i['properties']['language']
            found.update( {language: lemma} )
            row.append(found)
        for d in row:
            for key, value in d.items():
                dd[key].append(value)
        for lang in dd:
            s = ', '.join(dd[lang])
            dd[lang] = [s]
        total_list_of_results.append(dd)
f1.close()
```

```python
part5_ukredlinks_wiki = pd.DataFrame(total_list_of_results)
```

## Query Wikipedia Editions (as an additional information to create ground truth)

```python
red_titles1 = pd.read_csv(PATH_TO_DATA+"part2_ukredlinks_wiki.csv")
```

```python
red_titles1
```

```python
list_of_red_links1 = red_titles1['red_link_name'].tolist()
```

```python
# query en wikipedia
wiki_en_results = []
for title in list_of_red_links1:
    wiki_en_response = requests.get('https://en.wikipedia.org/wiki/' + title)
    wiki_str_response = str(wiki_en_response)
    if wiki_str_response == '<Response [200]>':
        wiki_en_results.append('True')
    elif wiki_str_response == '<Response [404]>':
        wiki_en_results.append('False')
    else:
        wiki_en_results.append('unknown')
```

```python
df_en_wiki = pd.DataFrame(wiki_en_results)
```

```python
with_enwiki = red_titles1.join(df_en_wiki)
```

```python
with_enwiki.rename(columns = {0:'en_wiki_res'}, inplace = True)
```
